**Exercise 3**

When I arrive to the coffee machine, I hear my colleague talking about the per-unit costs of producing 'product B' for the company. As the company produces more units, the per-unit costs continue to decrease until a point where they start to increase.


To optimize the per-unit production cost at its minimum to optimize efficiency, the company would need to find the number of units to be produced where the per-unit production costs begin to change from decreasing to increasing.


**Build a quadratic function $f(x) = 0.1(x)^2-9x+4500 \quad\text{on } x \in [0,100]$ to create the per-unit cost function, and make a conclusion.**

```python
def f(x):
    return 0.1*x**2 - 9*x + 4500

# Rango y punto Ã³ptimo
x = np.linspace(0, 100, 501)
x_opt = 45
y_opt = f(x_opt)

# Tabla cada 5 unidades
x_tab = np.arange(0, 101, 5)
df = pd.DataFrame({
    "unidades (x)": x_tab,
    "costo_unitario f(x)": f(x_tab).round(2)
})
print(df.head(10))
print("\nÃ“ptimo: x = {}, f(x) = {}".format(x_opt, y_opt))

# GrÃ¡fico
plt.plot(x, f(x), label="Costo unitario f(x)")
plt.scatter([x_opt], [y_opt], s=60)
plt.annotate(f"mÃ­nimo en x={x_opt}\nf(x)={y_opt:.2f}",
             (x_opt, y_opt), xytext=(x_opt+5, y_opt+30),
             arrowprops=dict(arrowstyle="->"))
plt.xlim(0, 100); plt.xlabel("Unidades producidas (x)")
plt.ylabel("Costo por unidad"); plt.grid(True); plt.legend()
plt.show()
```

**Exercise 1: Sum of two matrices**

Suppose we have two matrices A and B.

A = [[1,2],[3,4]]
B = [[4,5],[6,7]]

then we get
A+B = [[5,7],[9,11]]
A-B = [[-3,-3],[-3,-3]]

Make the sum of two matrices using Python with NumPy


```python
#DefiniciÃ³n de matrices

A = np.array([[1,2],
              [3,4]])

B = np.array([[4,5],
              [6,7]])

#Suma (C) y resta (D) de matrices

C = A + B

D = A - B

print("Matriz A:\n", A)

print("Matriz B:\n", B)

print("A + B:\n", C)

print("A - B:\n", D)
```

**Exercise 2: Sum of two lists**

There will be many situations in which we'll have to find an index-wise summation of two different lists. This can have possible applications in day-to-day programming. In this exercise, we will solve the same problem in various ways in which this task can be performed.

We have the following two lists:

list1 = [2, 5, 4, 7, 3]
list2 = [1, 4, 6, 9, 10]

Now let's use Python code to demonstrate addition of two lists.

```python
#MÃ©todo ingenuo

#DefiniciÃ³n de listas

list1 = [2,5,4,7,3]
list2 = [1,4,6,9,10]

print("Lista 1: " + str(list1))
print("Lista 2: " + str(list2))

res_list = []
for i in range(0, len(list1)):
  res_list.append(list1[i] + list2[i])

print("Lista resultado: " + str(res_list))
```

```python
#MÃ©todo de comprehension list

#DefiniciÃ³n de listas

list1 = [2,5,4,7,3]
list2 = [1,4,6,9,10]

lista_resultado = [list1 + list2 for list1, list2 in zip(list1, list2)]

print("Lista 1: ", list1)
print("Lista 2: ", list2)
print("Lista resultado: ", lista_resultado)
```

```python
#MÃ©todo Map().add()
from operator import add

list1 = [2,5,4,7,3]
list2 = [1,4,6,9,10]

lista_resultado = list(map(add,list1,list2))

print("Lista 1 + Lista 2 =", lista_resultado)
```

```python
#MÃ©todo zip().sum()
list1 = [2,5,4,7,3]
list2 = [1,4,6,9,10]

litsa_resultado = [sum(x) for x in zip(list1, list2)]

print("Lista  + Lista 2= ", lista_resultado)
```


**Exercise 3: Dot multiplication**

We have two matrices:

matrix1 = [[1,7,3],
 [4,5,2],
 [3,6,1]]
matrix2 = [[5,4,1],
 [1,2,3],
 [4,5,2]]

A simple technique but expensive method for larger input datasets is using for loops. In this exercise, we will first use nested for loops to iterate through each row and column of the matrices, and then we will perform the same multiplication using NumPy.


```python
matrix1 = np.array([[1,7,3],
           [4,5,2],
           [3,6,1]])

matrix2 = np.array([[5,4,1],
           [1,2,3],
           [4,5,2]])

# MÃ©todo 2: dot product
matrix_result = np.dot(matrix1, matrix2)    # equivalente a A @ B

print("Resultado del producto:\n", matrix_result)
```
